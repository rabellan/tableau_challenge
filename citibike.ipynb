## Citibike Data for Jersey City for October - December of 2023

```python
# Import dependencies
import pandas as pd
```

```python
# Create empty list to contain combined csv files
combined_csv_list = []

# List of Jersey City Citi Bike csv file names to be combined 
jc_csv_list = ['data/JC-202312-citibike-tripdata.csv', 'data/JC-202311-citibike-tripdata.csv', 'data/JC-202310-citibike-tripdata.csv']

# Loop through each file in the jc_csv_list
for file in jc_csv_list:
    # Read the file into a dataframe
    df_jc = pd.read_csv(file)

    # Append the dataframe to the combined_csv_list
    combined_csv_list.append(df_jc)

# Concatenate all dataframes in the list into one dataframe
df = pd.concat(combined_csv_list)

# Show dataframe
df
```

```python
# Save the dataframe to a CSV file
df.to_csv('data/citibike_oct_dec_2023.csv', index=False)
```
